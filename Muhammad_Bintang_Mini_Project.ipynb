```python
from sklearn.model_selection import StratifiedKFold
import numpy as np

# Fungsi untuk menghitung sensitivity, specificity, dan F1-score
def calculate_metrics(y_true, y_pred):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()

    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    f1_score = 2 * (sensitivity * specificity) / (sensitivity + specificity)

    return sensitivity, specificity, accuracy, f1_score

# Fungsi untuk 10-fold cross-validation
def k_fold_cross_validation(model, X, y):
    skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

    sensitivities = []
    specificities = []
    accuracies = []
    f1_scores = []

    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        sensitivity, specificity, accuracy, f1 = calculate_metrics(y_test, y_pred)

        sensitivities.append(sensitivity)
        specificities.append(specificity)
        accuracies.append(accuracy)
        f1_scores.append(f1)

    avg_sensitivity = np.mean(sensitivities)
    avg_specificity = np.mean(specificities)
    avg_accuracy = np.mean(accuracies)
    avg_f1_score = np.mean(f1_scores)

    return avg_sensitivity, avg_specificity, avg_accuracy, avg_f1_score

# Classifier yang digunakan: SVM
svm_classifier = SVC()

# Classifier yang digunakan: K-NN
knn_classifier = KNeighborsClassifier()

# Train and evaluate models for each file
for file_path in file_paths:
    print("File Path:", file_path)
    df = pd.read_excel(file_path)

    # Drop rows with missing values
    df = df.dropna()

    # Split features and target
    X = df.iloc[:, :-1]  # Ambil semua kolom kecuali kolom terakhir sebagai fitur
    y = df.iloc[:, -1]   # Ambil kolom terakhir sebagai target

    print("-------------- 10-fold Cross Validation --------------")
    # 10-fold Cross Validation with SVM
    avg_sensitivity_svm, avg_specificity_svm, avg_accuracy_svm, avg_f1_score_svm = k_fold_cross_validation(svm_classifier, X, y)
    print("SVM - Mean Sensitivity:", avg_sensitivity_svm)
    print("SVM - Mean Specificity:", avg_specificity_svm)
    print("SVM - Mean Accuracy:", avg_accuracy_svm)
    print("SVM - Mean F1-Score:", avg_f1_score_svm)

    # 10-fold Cross Validation with K-NN
    avg_sensitivity_knn, avg_specificity_knn, avg_accuracy_knn, avg_f1_score_knn = k_fold_cross_validation(knn_classifier, X, y)
    print("K-NN - Mean Sensitivity:", avg_sensitivity_knn)
    print("K-NN - Mean Specificity:", avg_specificity_knn)
    print("K-NN - Mean Accuracy:", avg_accuracy_knn)
    print("K-NN - Mean F1-Score:", avg_f1_score_knn)
    print("------------------------------------------------------")
```
